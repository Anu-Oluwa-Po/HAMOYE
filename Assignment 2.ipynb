```python
from sklearn.linear_model import LinearRegression

x = energy[['T2']]
y = energy['T6']


from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)

lr = LinearRegression()
lr.fit(x_train, y_train)
pred = lr.predict(x_test)
rmse = np.sqrt(mean_squared_error(y_test, pred))
round(rmse, 3)
```

```python
energy = energy.drop(['date', 'lights'], axis=1)
features = energy.drop(['Appliances'], axis=1)
target = energy['Appliances']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(features, target, test_size=0.3, random_state=42)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
```

```python
x_train = pd.DataFrame(scaler.fit_transform(x_train), columns=features.columns)
x_test = pd.DataFrame(scaler.fit_transform(x_test), columns=features.columns)
```

```python
x_train = scaler.transform(x_train)
x_test = scaler.transform(x_test)
```

```python
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

lr = LinearRegression()
lr.fit(x_train, y_train)
pred = lr.predict(x_train)
```

```python
mae = mean_absolute_error(y_train, pred)
round(mae, 3)
```

```python
rmse = np.sqrt(mean_squared_error(y_train, pred))
round(rmse, 3)
```

```python
pred = lr.predict(x_test)
mae= mean_absolute_error(y_test, pred)
round(mae, 3)
```

```python
pred = lr.predict(x_test)
rmse = np.sqrt(mean_squared_error(y_test, pred))
round(rmse, 3)
```

```python
ridge = Ridge()
ridge.fit(x_train, y_train)
```

```python
pred = ridge.predict(x_test)
np.sqrt(mean_squared_error(y_test, pred))
```

```python
lasso = Lasso()
lasso.fit(x_train, y_train)
```

```python
preds = lasso.predict(x_test)
np.sqrt(mean_squared_error(y_test, preds))
```

```python
def weights(model, features, column_name):
    weight = pd.Series(model.coef_, features.columns).sort_values()
    weight_df = pd.DataFrame(weight).reset_index()
    weight_df.columns = ["Features", column_name]
    weight_df[column_name].round(3)
    return weight_df
```

```python
lr_weights = weights(lr, features, 'LinearRegression')
ridge_w = weights(ridge, features, 'RidgeRegression')
lasso_w = weights(lasso, features, 'LassoRegression')

weight_data = pd.merge(lr_weights, ridge_w, on='Features')
weight_data = pd.merge(weight_data, lasso_w, on='Features')
```
